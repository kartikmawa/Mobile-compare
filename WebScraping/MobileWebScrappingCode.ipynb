```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
url = 'https://www.amazon.in/s?k=apple&rh=n%3A1389401031&ref=nb_sb_noss'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

page = requests.get(url,headers=headers)
response = requests.get(url, headers=headers)
soup1 = BeautifulSoup(page.content,'html.parser')
soup2 = BeautifulSoup(soup1.prettify(),'html.parser')
soup4 = BeautifulSoup(response.text, "html.parser")
```

```python
products = soup2.find_all("div", {"data-component-type": "s-search-result"})
products
```

```python
elements_with_data_index = soup2.find_all(attrs={"data-index": True})
count=len(elements_with_data_index)
count
```

```python
data_index_values = [element.get('data-index') for element in elements_with_data_index]
print("Data-index values found:", data_index_values)
```

```python
data = []
for product in products:
    # Thumbnail image
    img_tag = product.find("img", class_="s-image")
    img_url = img_tag["src"] if img_tag else None

    # Mobile name
    name_tag = product.find("h2", {"aria-label": True})
    name = name_tag["aria-label"].strip() if name_tag else None

    # Discounted price
    price_tag = product.find("span", class_="a-price")
    discounted_price = price_tag.find("span", class_="a-offscreen").text.strip() if price_tag else None

    # Actual price (M.R.P.)
    mrp_tag = product.find("span", class_="a-price a-text-price")
    actual_price = mrp_tag.find("span", class_="a-offscreen").text.strip() if mrp_tag else None

    # Rating
    rating_tag = product.find("span", class_="a-icon-alt")
    rating = rating_tag.text.strip() if rating_tag else None

    # Optional: remove multiple spaces inside text
    def clean_text(t):
        return re.sub(r'\s+', ' ', t).strip() if t else None

    data.append({
        "Image URL": clean_text(img_url),
        "Mobile Name": clean_text(name),
        "Discounted Price": clean_text(discounted_price),
        "Actual Price": clean_text(actual_price),
        "Rating": clean_text(rating)
    })

# Save into DataFrame
df = pd.DataFrame(data)
print(df)

df.to_excel("C:\\Users\\karti\\OneDrive\\Desktop\\MobComp\\Output_Products\\amazon_products.xlsx", index=False)

# with open("C:\\Users\\karti\\OneDrive\\Desktop\\MobComp\\Output_Products\\amazon_products.csv", "w") as f:
#     for p in data:
#         f.write(str(p))
```
